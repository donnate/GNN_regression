```python
##### Generate 
# Define the grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define the 2D Gaussian function
def gaussian(x, y, mu_x, mu_y, sigma_x, sigma_y):
    return np.exp(-((x - mu_x)**2 / (2 * sigma_x**2) + (y - mu_y)**2 / (2 * sigma_y**2)))

# Parameters for the Gaussian function
mu_x = 0
mu_y = 0
sigma_x = 3
sigma_y = 3


def gaussian(X, mu, sigma):
    dim_ambiant = X.shape[1]
    log_exp = 0
    for i in np.arange(dim_ambiant):
        log_exp += (X[:,i] - mu[i])**2 / (2 * sigma[i]**2)
        
    return np.exp(-log_exp)

mu = [0, 1]
sigma = [1,3]
X = np.hstack([x.reshape([-1,1]), y.reshape([-1,1])])
    
# Generate the 2D Gaussian function over the grid
z = gaussian(X, mu, sigma)

# Plot the function
plt.figure(figsize=(8, 6))
plt.contourf(x, y, z.reshape(x.shape), cmap='viridis')
plt.colorbar(label='Function Value')
plt.title('2D Gaussian Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Define the 2D sine and cosine functions
def sine_function(x, y, frequency_x, frequency_y):
    return np.sin(frequency_x * x) * np.sin(frequency_y * y)

def cosine_function(x, y, frequency_x, frequency_y):
    return np.cos(frequency_x * x) * np.cos(frequency_y * y)

# Parameters for the sine and cosine functions
frequency_x = 1
frequency_y = 1

# Generate the 2D sine and cosine functions over the grid
z_sine = sine_function(x, y, frequency_x, frequency_y)
z_cosine = cosine_function(x, y, frequency_x, frequency_y)

# Plot the sine function
plt.figure(figsize=(8, 6))
plt.contourf(x, y, z_sine, cmap='viridis')
plt.colorbar(label='Function Value')
plt.title('2D Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot the cosine function
plt.figure(figsize=(8, 6))
plt.contourf(x, y, z_cosine, cmap='viridis')
plt.colorbar(label='Function Value')
plt.title('2D Cosine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
